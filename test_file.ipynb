```python
import requests

def fetch_all_pages(api_url, items):
    page = 1
    all_results = []

    while True:
        response = requests.post(api_url, json={"items": items, "page": page})
        data = response.json()
        all_results.extend(data["classification_results"])

        if page >= data["total_pages"]:
            break

        page += 1

    return all_results

# Replace "your_api_url" with the actual URL of your API endpoint
api_url = "http://your-api-url/classify"
items = ['BAKERY', 'AN GRAIN BREAD', 'DELI', '...']  # Your list of items

all_results = fetch_all_pages(api_url, items)

# Now, 'all_results' contains the classification results from all pages
for result in all_results:
    print(result)
```

```python
import spacy
from typing import Any, Dict, List

import spacy
from typing import Any, Dict

def clean_strings_in_json(data: Dict[str, Any]) -> Dict[str, Any]:
    nlp = spacy.load("en_core_web_sm")
    cleaned_data = {}

    for key, value in data.items():
        if isinstance(value, str):
            doc = nlp(value)
            cleaned_text = ' '.join(token.text for token in doc if token.is_alpha)
            if cleaned_text.strip():  # Check if the cleaned text is not empty
                cleaned_data[key] = cleaned_text
        else:
            cleaned_data[key] = value

    return cleaned_data
#saved function EMERGENCY
```
