```python
# Shuffle the merged_data DataFrame
shuffled_data = merged_data.sample(frac=1).reset_index(drop=True)
shuffled_data.head()
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'shuffled_data' contains your shuffled DataFrame

# Splitting into train (70%), validation (15%), and test (15%)
train_data, test_val_data = train_test_split(shuffled_data, test_size=0.3, random_state=42)
val_data, test_data = train_test_split(test_val_data, test_size=0.5, random_state=42)

# Displaying the sizes of each set
print("Train set size:", len(train_data))
print("Validation set size:", len(val_data))
print("Test set size:", len(test_data))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import accuracy_score, classification_report



# Handling missing values in 'items' column (if necessary)
train_data['items'].fillna('', inplace=True)
val_data['items'].fillna('', inplace=True)
test_data['items'].fillna('', inplace=True)

# Separating features and labels
X_train_text, y_train = train_data['items'], train_data['label']
X_val_text, y_val = val_data['items'], val_data['label']
X_test_text, y_test = test_data['items'], test_data['label']

# Initialize TF-IDF Vectorizer for text data
vectorizer = TfidfVectorizer()  # Use TF-IDF instead of CountVectorizer
X_train = vectorizer.fit_transform(X_train_text)
X_val = vectorizer.transform(X_val_text)
X_test = vectorizer.transform(X_test_text)

# Initializing Logistic Regression model with regularization (L2) and cross-validation
model = LogisticRegressionCV(Cs=10, cv=5, penalty='l2', max_iter=1000)
model.fit(X_train, y_train)

# Predictions on validation data
val_predictions = model.predict(X_val)

# Evaluate the model on validation set
accuracy = accuracy_score(y_val, val_predictions)
print("Validation Accuracy:", accuracy)

# Predictions on test data
test_predictions = model.predict(X_test)

# Evaluate the model on test set
test_accuracy = accuracy_score(y_test, test_predictions)
print("Test Accuracy:", test_accuracy)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, test_predictions))
```

```python
import spacy
from typing import Any, Dict

def clean_strings_in_json(data: Dict[str, Any]) -> Dict[str, Any]:
    nlp = spacy.load("en_core_web_sm")
    metric_terms = {'lb', 'kg', 'cm', 'mm', 'gram', 'm', 'g', 'oz'}  # Add more metric terms as needed
    cleaned_data = {}

    for key, value in data.items():
        if isinstance(value, str):
            # Remove metric terms
            value = ' '.join(token.text for token in nlp(value) if token.text.lower() not in metric_terms)

            # Remove specific words
            value = value.replace('TAX', '').replace('TOTAL', '').replace('SUBTOTAL', '')
            value = value.replace('CARD', '').replace('DEBIT', '').replace('CREDIT', '')
            value = value.replace('Debit', '').replace('credit', '').replace('card', '')
            value = value.replace('Card', '').replace('k', '').replace('FOR', '')
            value = value.replace('for', '').replace('Welocme', '').replace('welcome', '')
            value = value.replace('thankyou', '').replace('Thankyou', '').replace('THNAKYOU', '')
            value = value.replace('balance', '').replace('BALANCE', '').replace('due', '')

            # Remove non-alphabetic characters
            cleaned_text = ' '.join(token.text for token in nlp(value) if token.is_alpha)

            if cleaned_text.strip():  # Check if the cleaned text is not empty or just spaces
                cleaned_data[key] = cleaned_text
        else:
            cleaned_data[key] = value

    return cleaned_data


```

```python
from fastapi import FastAPI, HTTPException
from typing import List
from pydantic import BaseModel


# FastAPI app instance
app = FastAPI()

# Pydantic model for request body
class ItemsRequest(BaseModel):
    items: List[str]

# Pydantic model for response
class ItemsResponse(BaseModel):
    classification_results: List[str]

# Endpoint to classify multiple items
@app.post("/classify")
async def classify(request: ItemsRequest):
    try:
        results = await classify_items(request.items)
        return ItemsResponse(classification_results=results['classification_results'])
    except Exception as e:
        raise HTTPException(status_code=500, detail=f"Error processing items: {str(e)}")

# endpoint for cleaning strings in JSON data
@app.post("/clean_strings")
async def clean_strings(data: dict):
    cleaned_data = clean_strings_in_json(data)
    return cleaned_data

```
